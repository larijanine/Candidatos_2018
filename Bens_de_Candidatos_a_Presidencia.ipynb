```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
file = 'Bens_candidatos_a_presidencia.xlsx'
dataframe_excel = pd.read_excel(file, sheet_name = 0, header = 0, index_col = 0)
```

**Dados Extraidos do portal do TSE**, do
site http://divulgacandcontas.tse.jus.br/divulga/#/estados/2018/2022802018/BR/candidatos, analisando todos os bens declarados dos candidatos a presidencia do ano de 2018.

```python
df = dataframe_excel.head(2000)
dataframe_excel.head(2000) 
```

**Com base nas informações, demonstra-se de forma gráfica o total em bens entre os candidatos.**

```python
df.groupby('CANDIDATOS')['VALOR DO BEM'].sum().plot.bar()
```

# Conclusão da análise

Conclui-se que entre os candidatos há desde quem não possui nenhum bem declarado, como o Cabo Daciolo, até candidatos milionários, como João Amoêdo e Henrique Meirelles, que fizeram sua fortuna no setor privado.
